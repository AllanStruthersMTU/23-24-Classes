# Solving ODEs
We are going to learn how to solve and plot an ODE
$$y' = f(t,y)$$
in Julia.

##### Warning
Installing the package took quite a long time. 
Precompiling the package took quite a long time.
There is a LOT of stuff in this package. 

```julia
import Pkg
Pkg.add("DifferentialEquations")
```

```julia
using DifferentialEquations
# Define a scalar test function
# p indicates parameters
function f(u,p,t) 
    sin((t+u)^2)
end
u0 = 1.2
tspan = (0.0,4.0)
ivp = ODEProblem(f,u0,tspan)
sol = solve(ivp,Tsit5())
```

It also replaced/update the plots package. 

```julia
using Plots
plot(sol)
```
